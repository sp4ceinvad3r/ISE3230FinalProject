```python
#Code for calculating single scenario with 24 hours
import cvxpy as cp

x = cp.Variable(10, boolean = True) # vector variable

constraints = []
constraints.append (4*x[0]+x[1]+5*x[2]+4*x[3]+3*x[4]+2*x [5]+3*x [6] +x[7]+2*x [8] +6*x [9]<=24)
obj_func= 0.456+.06*x[0]+0.02*x[1]+0.07*x[2]+0.05*x[3]+0.05*x[4]+0.01*x[5]+0.04*x[6]+0.02*x[7]+0.01*x[8]+0.07*x[9]

problem = cp.Problem (cp.Maximize(obj_func), constraints)
#problem.solve(verbose = True)
#problem.solve(solver=cp.CVXOPT, verbose = True)
problem.solve(solver=cp.GUROBI, verbose = True)
```
